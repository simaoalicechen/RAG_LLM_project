```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin


# Get PDFs from one website
def get_pdfs_from_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    pdf_links = []
    for link in soup.find_all('a', href=True):
        absolute_url = urljoin(url, link['href'])
        if absolute_url.endswith('.pdf'):
            pdf_links.append(absolute_url)

    return pdf_links

website_url = 'https://www.volpe.dot.gov/project-delivery/toolbox/public-engagement'
pdf_links = get_pdfs_from_website(website_url)

for pdf_link in pdf_links:
    print(f"Found PDF: {pdf_link}")
```

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def download_pdf(pdf_url, save_path):
    response = requests.get(pdf_url)
    with open(save_path, 'wb') as pdf_file:
        pdf_file.write(response.content)

# Recursively visit links from all web links
def get_links_and_pdfs(url, visited_urls=set()):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        links = [urljoin(url, a['href']) for a in soup.find_all('a', href=True)]
        pdfs = [urljoin(url, a['href']) for a in soup.find_all('a', href=True) if a['href'].lower().endswith('.pdf')]

        for pdf_url in pdfs:
            pdf_filename = os.path.basename(urlparse(pdf_url).path)
            download_pdf(pdf_url, pdf_filename)
            print(f"Downloaded PDF: {pdf_filename}")
        for link in links:
            if link not in visited_urls:
                visited_urls.add(link)
                get_links_and_pdfs(link, visited_urls)

    except requests.exceptions.RequestException as e:
        print(f"Error processing {url}: {e}")

# Example usage
start_url = 'https://www.volpe.dot.gov/project-delivery-toolbox'
get_links_and_pdfs(start_url)
```
